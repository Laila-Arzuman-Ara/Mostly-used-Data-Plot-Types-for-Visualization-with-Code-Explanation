Two common data plot types for visualization are scatter plots and histograms. Let's walk through the process of creating these plots in Python, using built-in datasets from the seaborn library.
First, let's import the necessary libraries and load in the dataset:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load in built-in dataset
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Add labels
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.title("Total Bill Amount vs. Tip Amount")
plt.show()
```

Scatter Plot
A scatter plot is used to display the relationship between two variables. In this example, we'll plot the relationship between the total bill amount and the tip amount in the tips dataset.

Explanation:

1.sns.scatterplot() creates the scatter plot. 
2.We specify the x-axis variable with x="total_bill", the y-axis variable with y="tip", and the dataset with data=tips.
3.plt.xlabel(), plt.ylabel(), and plt.title() add axis labels and a title to the plot.
4.plt.show() displays the plot.

```python
# Create histogram
sns.histplot(data=tips, x="total_bill")

# Add labels
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill Amounts")
plt.show()
```

Histogram
A histogram is used to visualize the distribution of a single variable. In this example, we'll plot the distribution of total bill amounts in the tips dataset.
Explanation:

1.sns.histplot() creates the histogram. We specify the variable with x="total_bill" and the dataset with data=tips.
2.plt.xlabel(), plt.ylabel(), and plt.title() add axis labels and a title to the plot.
3.plt.show() displays the plot.

That's it! These are just two examples of data plots that you can create in Python. With the right data and tools, the possibilities are endless.
